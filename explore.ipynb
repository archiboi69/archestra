```python
import geopandas as gpd

buildings = gpd.read_file("data/GIS/PL.PZGiK.336.BDOT10k.2261__OT_BUBD_A.gpkg")
buildings.to_crs(epsg=2177, inplace=True)

buildings.head()
```

```python
plots = gpd.read_file("data/GIS/EGIB/2261_dzialki_egib_wfs_gml.gml")
plots.set_crs(epsg=2177)

plots.head()
```

```python
roads = gpd.read_file('data/GIS/EGIB/2261_ulice_egib_wfs_gml.gml')
roads.set_crs(epsg=2177)

roads.head()
```

```python
road_plots = gpd.sjoin(plots, roads, predicate='intersects', lsuffix='plots', rsuffix='roads')

road_plots.head()
```
